```python
df['TotalCharges']=df['TotalCharges'].replace("","0")
```

replacing blanks in totalcharges with 0 and converting it into float

```python
# df['TotalCharges'] = df['TotalCharges'].str.strip()
df['TotalCharges'] = df['TotalCharges'].replace(" ", "0")
df['TotalCharges'] = df['TotalCharges'].astype("float")
```
